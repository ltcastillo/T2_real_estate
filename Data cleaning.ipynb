```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
file_path = Path('zillow_BayArea_ZipCodes.csv')
value_df = pd.read_csv(file_path)
value_df.head()
```

```python
value_df1=value_df[["regionname","city","8/31/2015","8/31/2016","8/31/2017","8/31/2018","8/31/2019","8/31/2020"]]
value_df1.head()
```

```python
value_df1.rename(columns={"8/31/2015": "08312015", "8/31/2016": "08312016","8/31/2017": "08312017", 
                          "8/31/2018": "08312018","8/31/2019": "08312019", "8/31/2020": "08312020"})
```

```python
value_df2=value_df1.dropna()
```

```python
value_df2.dtypes
```

```python
value_df2['202008']=value_df2['8/31/2020']/value_df2['8/31/2019']-1
value_df2['201908']=value_df2['8/31/2019']/value_df2['8/31/2018']-1
value_df2['201808']=value_df2['8/31/2018']/value_df2['8/31/2017']-1
value_df2['201708']=value_df2['8/31/2017']/value_df2['8/31/2016']-1
value_df2['201608']=value_df2['8/31/2016']/value_df2['8/31/2015']-1
value_df2.head()
```

```python
value_df3=value_df2.drop(columns=['8/31/2015', '8/31/2016','8/31/2017', '8/31/2018','8/31/2019', '8/31/2020'])
value_df3.head()
```

```python
value_df3.to_csv('Value_Incr', index=False)
```

```python
value_df4=value_df3.drop(columns=["201908","201808","201708","201608"])
value_df4['date']='202008'
value_df4=value_df4.rename(columns={"202008": "Incr_Per"})
value_df4.head()
```

```python
value_df5=value_df3.drop(columns=["202008","201808","201708","201608"])
value_df5['date']='201908'
value_df5=value_df5.rename(columns={"201908": "Incr_Per"})

value_df6=value_df3.drop(columns=["201908","202008","201708","201608"])
value_df6['date']='201808'
value_df6=value_df6.rename(columns={"201808": "Incr_Per"})

value_df7=value_df3.drop(columns=["201908","201808","202008","201608"])
value_df7['date']='201708'
value_df7=value_df7.rename(columns={"201708": "Incr_Per"})

value_df8=value_df3.drop(columns=["201908","201808","201708","202008"])
value_df8['date']='201608'
value_df8=value_df8.rename(columns={"201608": "Incr_Per"})
value_df8.head()
```

```python
value_df7.head()
```

```python
value_df6.head()
```

```python
value_df5.head()
```

```python
frames=[value_df4,value_df5,value_df6,value_df7,value_df8]
value_incr=pd.concat(frames)
```

```python
value_incr['Incr_type']='NA'
value_incr.loc[(value_incr['Incr_Per']>0.1), 'Incr_type'] = "Huge_Incr"
value_incr.loc[(value_incr['Incr_Per']<=0.1)&(value_incr['Incr_Per']>0.07), 'Incr_type'] = "High_Incr"
value_incr.loc[(value_incr['Incr_Per']<=0.07)&(value_incr['Incr_Per']>0.05), 'Incr_type'] = "Moderate_Incr"
value_incr.loc[(value_incr['Incr_Per']<=0.05)&(value_incr['Incr_Per']>0.03), 'Incr_type'] = "Small_Incr"
value_incr.loc[(value_incr['Incr_Per']<=0.03)&(value_incr['Incr_Per']>0), 'Incr_type'] = "Stable_Incr"
value_incr.loc[(value_incr['Incr_Per']<=0)&(value_incr['Incr_Per']>-0.05), 'Incr_type'] = "Small_Decr"
value_incr.loc[(value_incr['Incr_Per']<=-0.05)&(value_incr['Incr_Per']>-0.07), 'Incr_type'] = "Moderate_Decr"
value_incr.loc[(value_incr['Incr_Per']<=-0.07)&(value_incr['Incr_Per']>-0.1), 'Incr_type'] = "High_Decr"
value_incr.loc[(value_incr['Incr_Per']<-0.1), 'Incr_type'] = "Huge_Decr"
value_incr.head()
```

```python
value_incr['Year']='0'
value_incr.loc[(value_incr['date']=='201608'), 'Year'] = 2016
value_incr.loc[(value_incr['date']=='201708'), 'Year'] = 2017
value_incr.loc[(value_incr['date']=='201808'), 'Year'] = 2018
value_incr.loc[(value_incr['date']=='201908'), 'Year'] = 2019
value_incr.loc[(value_incr['date']=='202008'), 'Year'] = 2020
value_incr.head()
```

```python
value_incr.to_csv('Value_Incr_Combined', index=False)
```
