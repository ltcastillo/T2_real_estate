```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
file_path = Path('realtordotcom_core.csv')
core_df = pd.read_csv(file_path)
core_df.head()
```

```python
core_df1=core_df[["month_date_yyyymm","postal_code","zip_name","active_listing_count_yy","median_days_on_market_yy","new_listing_count_yy","pending_ratio_yy"]]
core_df1.head()
```

```python
core_df2=core_df1.dropna()
core_df2.head()
```

```python
core_df3=core_df2.loc[((core_df2['month_date_yyyymm']==202007) |(core_df2['month_date_yyyymm']==201907) |(core_df2['month_date_yyyymm'] ==201807)
                      | (core_df2['month_date_yyyymm']==201707) |(core_df2['month_date_yyyymm']==201607))]
core_df3.head()
```

```python
core_df3['Year']='0'
core_df3.loc[(core_df3['month_date_yyyymm']==201607), 'Year'] = 2016
core_df3.loc[(core_df3['month_date_yyyymm']==201707), 'Year'] = 2017
core_df3.loc[(core_df3['month_date_yyyymm']==201807), 'Year'] = 2018
core_df3.loc[(core_df3['month_date_yyyymm']==201907), 'Year'] = 2019
core_df3.loc[(core_df3['month_date_yyyymm']==202007), 'Year'] = 2020
core_df3.head()
```

```python
core_df3.to_csv('core', index=False)
```
